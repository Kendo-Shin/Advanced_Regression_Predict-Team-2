```python
# remove missing values/ features

#23% valencia pressure is missing
#set valencia pressure to mode
mode_value=df_train['Valencia_pressure'].mode()

df_train['Valencia_pressure'] = df_train['Valencia_pressure'].replace(np.nan, mode_value[0])

df_test['Valencia_pressure'] = df_test['Valencia_pressure'].replace(np.nan, mode_value[0])
```

```python
#i'm assuming time of day affects how energy is used, so theres a corelation between time of day and

df_train['time_of_day']=df_train['time'].apply(lambda x: x.split()[1].split(':')[0]) #return on time of day only
df_train['month_of_yr']=df_train['time'].apply(lambda x: x.split()[0].split('-')[1]) #return on momth of yr only
df_train['yr']=df_train['time'].apply(lambda x: x.split()[0].split('-')[0]) #return on yr only


df_test['time_of_day']=df_test['time'].apply(lambda x: x.split()[1].split(':')[0])
df_test['month_of_yr']=df_test['time'].apply(lambda x: x.split()[0].split('-')[1])
df_test['yr']=df_test['time'].apply(lambda x: x.split()[0].split('-')[0])


#during the early hrs of the morning theres a very low short fall on average.
#midday and mid midnight we see a hike in short short fall

```

```python

df_train.drop(['Unnamed: 0', 'time'], axis=1, inplace=True)


time=df_test['time']  ##will be needed for submission
df_test.drop(['Unnamed: 0', 'time'], axis=1, inplace=True)
```

```python
#convert valencia wind deg to numerical values
df_train['Valencia_wind_deg']=df_train['Valencia_wind_deg'].apply(lambda x: x.split('_')[1])
    
df_test['Valencia_wind_deg']=df_test['Valencia_wind_deg'].apply(lambda x: x.split('_')[1])  
```

```python
df_train['Seville_pressure']=df_train['Seville_pressure'].apply(lambda x: x[2:])

df_test['Seville_pressure']=df_test['Seville_pressure'].apply(lambda x: x[2:])
```

```python
# split data
## Splitting our data into dependent Variable and Independent Variable
X = df_train.drop(columns = 'load_shortfall_3h')
Y = df_train['load_shortfall_3h']

```

```python
# create targets and features dataset

scale = StandardScaler()
X = scale.fit_transform(X)

```

```python
# create one or more ML models

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 42)
```

```python
# evaluate one or more ML models
#my group models have already decided that we will use the random forest

# Our forest consists of 200 trees with a max depth of 8 
RF = RandomForestRegressor(n_estimators=200, max_depth=8)
# Fitting the model
RF.fit(X_train,Y_train)
RF_predict = RF.predict(X_test)

```
