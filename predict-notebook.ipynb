### 4.1 Base Model Data Engineering
For purposes of the creating the base model, we will drop all non-numeric columns and columns with null values.

As outlined in section 3.1, the non-numeric columns are:
- time
- Valencia_wind_deg
- Seville_pressure

The column that contains null values is:
- Valencia_pressure

We will create a `drop_columns` function that will facilitate this process.

---

```python
# Function to drop non-numeric columns and columns with null values
def drop_columns(input_df):
    output_df = input_df.copy()
    for column in output_df:
        with_nulls = output_df[column].isna().sum()
        if with_nulls > 0:
            output_df = output_df.drop(column, axis = 1)
        else:
            if column not in output_df.select_dtypes(include='number').columns:
                output_df = output_df.drop(column, axis = 1)
    return output_df
```

We will call our function on our training and test datasets.

```python
# The new training dataset
base_clean_df_train = drop_columns(df_train)

# The new test dataset
base_clean_df_test = drop_columns(df_test)
```

```python
# Confirm redundancy

df_train['Unnamed: 0'].head()
```

From the results above, we confirm that the values of `Unnamed: 0` column matches the index.

We will drop the column using a `drop_unnamed` function.

```python
# Function to drop "Unnamed: 0" column

def drop_unnamed(input_df):
    output_df = input_df.copy()
    output_df = output_df.drop(['Unnamed: 0'], axis = 1)
    return output_df
```

Drop the column in both the training and testing datasets.

```python
# Drop "Unnamed: 0"

df_train_no_unnamed = drop_unnamed(df_train)
df_test_no_unnamed = drop_unnamed(df_test)

#Drop confirmation
display(df_train_no_unnamed.sample(2))
display(df_test_no_unnamed.sample(2))
```

#### time

`time` column is currently a string, that is, object type. We will convert it to the datetime format, then create 4 new columns from it namely: `hour`, `day`, `month` and `year`. We chose to extract `hour` because the data is periodic, that is, every 3 hours. The `day` and `month` will help in capturing the seasonal changes, and the `year` will reflect the changes over the years. 

We will then drop the `time` column, and we will do this using function `process_date`.

```python
# Function to process date

def process_date(input_df):
    output_df = input_df.copy()
    output_df['time'] = pd.to_datetime(output_df['time'])
    output_df['hour'] = output_df['time'].dt.hour
    output_df['day'] = output_df['time'].dt.day
    output_df['month'] = output_df['time'].dt.month
    output_df['year'] = output_df['time'].dt.year
    output_df = output_df.drop(['time'], axis = 1)
    return output_df  
```

Process the dates in both the training and testing datasets.

```python
# Process date
df_train_date_processed = process_date(df_train_no_unnamed)
df_test_date_processed = process_date(df_test_no_unnamed)

#Date processing confirmation
display(df_train_date_processed.sample(2))
display(df_test_date_processed.sample(2))
```

#### Valencia_wind_deg

We will start by comparing the values of `Valencia_wind_deg` column with the values of other cities wind degree columns.

```python
# Comparing {city}_wind_deg columns

df_train_wind_deg = df_train_date_processed[['Valencia_wind_deg', 'Bilbao_wind_deg', 'Barcelona_wind_deg']]
df_train_wind_deg.head()
```

From the above results, `Valencia_wind_deg` has been expressed categorically whereas `Bilbao_wind_deg` and `Barcelona_wind_deg` are actual values.

Next, we will look at the values used in expressing it.

```python
# Generate list of unique values used

wind_deg_values = list(df_train_date_processed['Valencia_wind_deg'].unique())
print(wind_deg_values)
```

From the results above, we can see that `Valencia_wind_deg` has been expressed as a level between 1 and 10. However, we do not know the scale that has been used to perform the categorization therefore we cannot use the values in `Bilbao_wind_deg` and `Barcelona_wind_deg`to get corresponding values.

We will therefore perform one-hot encoding to ensure that the levels are not weighted improperly. After performing the encoding, we will drop the `Valencia_wind_deg` column.

```python
# Function to encode Valencia_wind_deg

def dummy_encode_valencia_wind_deg(input_df):
    output_df = input_df.copy()
    vwd_dummies = pd.get_dummies(output_df['Valencia_wind_deg'],
                                 drop_first = True).rename(columns = lambda x: 'Valencia_wind_deg_' + str(x))
    output_df = pd.concat([output_df, vwd_dummies], axis = 1)
    output_df = output_df.drop(['Valencia_wind_deg'], axis = 1)
    return output_df
```

Encode `Valencia_wind_deg` values.

```python
# Encoding
df_train_valencia_wd_encoded = dummy_encode_valencia_wind_deg(df_train_date_processed)
df_test_valencia_wd_encoded = dummy_encode_valencia_wind_deg(df_test_date_processed)

#Encoding confirmation
display(df_train_valencia_wd_encoded.sample(2))
display(df_test_valencia_wd_encoded.sample(2))
```

Next, we will look at the values used in expressing it.

```python
# Generate list of unique values used

pressure_values = list(df_train_valencia_wd_encoded['Seville_pressure'].unique())
print(pressure_values)
```

We will therefore perform one-hot encoding to ensure that the categories are not weighted improperly. After performing the encoding, we will drop the `Seville_pressure` column.

```python
# Function to encode Seville_pressure

def dummy_encode_seville_pressure(input_df):
    output_df = input_df.copy()
    sp_dummies = pd.get_dummies(output_df['Seville_pressure'],
                                 drop_first = True).rename(columns = lambda x: 'Seville_pressure_' + str(x))
    output_df = pd.concat([output_df, sp_dummies], axis = 1)
    output_df = output_df.drop(['Seville_pressure'], axis = 1)
    return output_df
```

Encode `Seville_pressure` values.

```python
# Encoding
df_train_seville_p_encoded = dummy_encode_seville_pressure(df_train_valencia_wd_encoded)
df_test_seville_p_encoded = dummy_encode_seville_pressure(df_test_valencia_wd_encoded)

#Encoding confirmation
display(df_train_seville_p_encoded.sample(2))
display(df_test_seville_p_encoded.sample(2))
```
