### Statistical Analysis:

---


`df.describe()` will be used to get the count, mean, standard deviation and a five number summary of each numerical feature.

The five number summary include: (Minimum, Lower Quartile (Q1) = 25%, Median (Q2) = 50%, Upper Quartile (Q3) = 75%, Maximum).


```python
# look at data statistics
description = df.describe().T
description.style.background_gradient(cmap = 'BrBG')  
```

---
`df.describe()` can also be used to get some statistical values from object-type features. Specifying the argument: include = 'O' which stands for object, gives us a result which will include count, unique, top, and freq.

The unique is the number of unique values in the feature.

The top is the most common value.

The freq is the most common value’s frequency.

```python
# check statistics on object-type features
df.describe(include='O')
```

`df.kurtosis()` will be used to measure outliers present in the data.

**High kurtosis (>3)** indicates a large number of outliers and **low kurtosis (<3)** a lack of outliers.

```python
# checking the kurtosis of the df
df.kurtosis(numeric_only = True)
```

---
Looking at the output above, we can easily see that the `Valencia_snow_3h` and `Barcelona_pressure` columns contains significantly huge amounts of outliers followed by `Bilbao_snow_3h`, `Seville_rain_3h`, `Barcelona_rain_3h` and `Barcelona_rain_1h`.

  
`df.skew()` will be used to indicate how symmetrical the data is. Below is a table that explains the range of values with regards to skew.

|   Skew Value (x)  |       Description of Data      |
|:-------------------|:---------------:|
| -0.5 < x < 0.5              |Fairly Symmetrical |
| -1 < x < -0.5 | Moderate Negative Skew  | 
| 0.5 < x < 1             | Moderate Positive Skew  | 
|       x < -1     |High Negative Skew  | 
|       x > 1  |High Positve Skew | 


```python
# Checking the skew of the df
df.skew(numeric_only = True)
```

  
Looking at the output above, we can observe that several columns are fairly symetrical while `Valencia_snow_3h` and `Barcelona_pressure` had very high positive skews of values above 50 implicating that majority of the data for these columns are majorly on the greater positive side as compared to the negative side.

### Multivariate Analysis:  

We will employ this type of analysis to determine the relationship between any two numerical features by calculating their correlation coefficient. Correlation is defined as a measure of the degree to which two variables change together, if at all. If two features have a strong positive correlation, it means that if the value of one feature increases, the value of the other feature also increases. In this project we will apply the Pearson correlation which measures the linear relationship between features and assumes that the features are normally distributed. Below is a table that explains how to interpret the Pearson correlation measure:


|   Pearson Correlation Coefficient (r)  |       Description of Relationship     |
|:-------------------|:---------------:|
|  r = -1              |Perfect Negative Correlation |
| -1 < r < -0.8 | Strong Negative Correlation  | 
| - 0.8 < r < -0.5             | Moderate Negative Correlation  | 
|       - 0.5 < r < 0     |Weak Negative Correlation  | 
|       r = 0  |No Linear Correlation | 
| 0 < r < 0.5 | Weak Positive Correlation  | 
| 0.5 < r < 0.8             | Moderate Positive Correlation  | 
|       0.8 < r < 1     |Strong Positive Correlation  | 
|       r = 1  |Perfect Positive Correlation | 

```python
# evaluate correlation
correlation = df.corr()
correlation.style.background_gradient(cmap = 'BrBG')
```

  
Looking at the correlation output above, we notice that there were several moderate negative and positive correlations with correlation values within the ranges |0.5 < r < 0.8|. There were additonal strong positive and negative correlations observed with correlation values within the range |0.8 < r < 1|. 

We will inspect the `Valencia_pressure` feature so as to know how to handle the `null` values in the Data engineering phase.

```python
# looking at the distribution of Valencia_pressure
sns.displot(data= df['Valencia_pressure'], aspect= 3)
```

  
The plot shows how the values for Valencia_pressure are positively skewed with majority of the values ranging between 1010 and 1020. For this reason we will use the `Mode` to fill in the null values for this feature because the mode is a better representative of the missing values as compared to the mean whose value will be greatly affected by the high populated null values. 

```python
# Heat map
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```
